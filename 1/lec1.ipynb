## RAM машина
- чтение/запись/инициализация ячейки памяти стоит одну операцию
- можно выполнять арифметические/логические оперции с одной ячейкой памяти, а так же переходы
- похожа на компьютер

## Сложность алгоритма
Сложность алгоритма -- порядок количества действий зависимости от размера входных данных

Часто применяют О нотацию -- нестрогая оценка сверху

Суть: хотим понимать во сколько раз замедлится наша программа при увеличении размера входных данных
